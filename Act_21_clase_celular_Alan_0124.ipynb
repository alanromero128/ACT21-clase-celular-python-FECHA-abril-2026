```python
class Celular:
    # El constructor define los atributos iniciales
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    # Método para mostrar información
    def mostrar_info(self):
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Precio: ${self.precio}")
    def encender(self):
        self.encendido = True
        print("Celular encendido")


# --- Uso de la clase ---
# creacion de objetos(instancias)
celular1 = Celular("Samsung", "Galaxy S21", 800)
celular2 = Celular("Apple", "iPhone 13", 950)
# llamadas a las funciones
print("Información del Celular 1:")
celular1.mostrar_info()
celular1.encender()
print("-----")

print("Información del Celular 2:")
celular2.mostrar_info()
celular2.encender()
print("-----")

print("programa terminado ")
print("Creador de software por Alan Romero NL = 47")
```
